```python
df1 = df[['tweet', 'replies_count', 'retweets_count', 'likes_count', 'hashtags', 'cashtags', 'language']]
```

```python
df1.retweets_count = df1.retweets_count.astype(int)
df1.replies_count = df1.replies_count.astype(int)
df1.likes_count = df1.likes_count.astype(int)
```

```python
df1.dtypes
```

```python
df1 = df1.sort_values(['likes_count'], ascending=[False])
```

```python
from textblob import TextBlob
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", tweet).split())
```

```python
df1.tweet = df.tweet.astype('string')
```

```python
df1.tweet = df1.tweet.apply(clean_tweet)
```

```python
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
df1['polarity'] = df1.tweet.apply(getPolarity)
```

```python
df1.head(10)
```

```python
ntwts = df1.tweet[df1.polarity < 0]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def getPolarityVader(text):
    return analyzer.polarity_scores(text)
```

```python
df1['polarityVader'] = df1.tweet.apply(getPolarityVader)
```

```python
df1.head(10)
```

```python
def getCompond(x):
    return x['compound']
```

```python
df1['polarutyVaderComp'] = df1.polarityVader.apply(getCompond)
```

```python
df1.head(10)
```

```python
nvtwts = df1.tweet[df1.polarutyVaderComp < 0]
```
